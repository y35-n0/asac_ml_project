```python
import pandas as pd
```

```python
preprocessed = pd.read_csv('../data/preprocessed/samsung_preprocess.csv',
                           encoding='cp949',
                           index_col=0)
preprocessed = preprocessed.reset_index()

marcap_n = pd.read_csv('../data/origin/krx/삼성전자_marcap.csv',
                      encoding='cp949')

marcap_f = pd.read_csv('../data/origin/krx/삼성전자_marcap.csv',
                        encoding='cp949')
```

```python
preprocessed.columns
```

```python
df_n = pd.merge(preprocessed.loc[preprocessed['특정증권등의 종류_보통주'] == 1,['index', '날짜']],
         marcap_n[['Date', 'Adj Close', 'Stocks']],
         how='left',
         left_on='날짜',
         right_on='Date'
)
df_n = df_n.drop(columns = ['Date'])

df_f = pd.merge(preprocessed.loc[preprocessed['특정증권등의 종류_우선주'] == 1,['index', '날짜']],
         marcap_n[['Date', 'Adj Close', 'Stocks']],
         how='left',
         left_on='날짜',
         right_on='Date'
)
df_f = df_f.drop(columns=['Date'])

df = pd.concat([df_f, df_n])
df = df.drop(columns='날짜')
```

```python
result = pd.merge(preprocessed, df, on='index')
result = result.drop(columns=['index', 'Adj Close'])
result = result.rename(columns={'Stocks':'상장주식수'})
result
```

```python
result.to_csv('../data/preprocessed/20150101-20221024_삼성전자_d002_with_stocks.csv',
              encoding='cp949',
              index=False)
```

```python
result.columns
```

```python
use_cols = ['거래량', '개인','기관', '외인수량', '외국계', '프로그램', '변동전', '증감']
result_df = result.copy()
for col in use_cols:
    result_df[col] = result_df.apply(lambda x: x[col]/x['상장주식수']*100, axis=1)
    
result_df[use_cols]
```

```python
result_df.to_csv('../data/preprocessed/20150101-20221024_삼성전자_d002_with_stocks_ratio.csv',
              encoding='cp949',
              index=False)
```
